Utilize os arquivos livros.csv, livros_genero.csv e livros_classicos.csv
como entrada para o processamento requerido nos seguintes exercícios.
1) Faça a concatenação das bases de dados e encontre qual o gênero de livro
correspondente a cada livro.

2) Faça a concatenação das bases de dados de livros e livros_genero e
encontre qual os gêneros e os autores que não possuem um livro

3) Faça a junção das bases de livros e livros clássicos.

4) Faça a junção das bases de livros e livros clássicos apenas com os

dataframes.

5) Faça a junção das bases de livros e livros clássicos apenas com os
elementos comuns entre ambos, e adicione chaves para o dataframe de
livros e de livros clássicos.

6) Faça a junção das bases de livros e livros clássicos apenas com os
elementos comuns entre ambos, e adicione chaves para o dataframe de

7) Faça a junção das bases de dados livros e livros_classicos e verifique a
integridade do dataframe resultante

## ***Bibliotecas***

```python
import pandas as pd
```

Abrindo arquivos.

```python
livros = pd.read_csv('livros.csv', sep=',')
livros.head()
```

```python
livros_classicos = pd.read_csv('livros_classicos.csv', sep=',')
livros_classicos.head()
```

```python
livros_genero = pd.read_csv('livros_genero.csv', sep=',')
livros_genero.head()
```

# **Exercícios**

1) Faça a concatenação das bases de dados e encontre qual o gênero de livro correspondente a cada livro.

```python
pd.concat([livros, livros_genero], axis=1)
```

2) Faça a concatenação das bases de dados de livros e livros_genero e encontre qual os gêneros e os autores que não possuem um livro correspondente.

```python
tabela_1 = pd.concat([livros_genero,livros], axis = 1)
mask = tabela_1['nome'].isnull()
registro = tabela_1[mask]

for index,reg in registro.iterrows():
    print(f'O autor {reg["autor"]} do genero {reg["genero"]} não possui um livro correspondente.')   
```

3) Faça a junção das bases de livros e livros clássicos.

```python
pd.concat([livros, livros_classicos], axis = 1)
```

4) Faça a junção das bases de livros e livros clássicos apenas com os elementos em comum entre ambos e remova os índices dos antigos dataframes.

```python
pd.concat([livros, livros_classicos], axis = 1, join = 'inner', ignore_index = True)
```

5) Faça a junção das bases de livros e livros clássicos apenas com os elementos comuns entre ambos, e adicione chaves para o dataframe de livros e de livros clássicos.

```python
pd.concat([livros, livros_classicos], axis = 1, join = 'inner', keys = ['livros', 'livros_classicos'])
```

6) Faça a junção das bases de livros e livros clássicos apenas com os elementos comuns entre ambos, e adicione chaves para o dataframe de livros e de livros clássicos, além de nomes para os índices.

```python
pd.concat([livros, livros_classicos], axis = 1, join = 'inner', keys = ['livros', 'livros_classicos'], names = ['tradicionais', 'classicos'])
```

7) Faça a junção das bases de dados livros e livros_classicos e verifique a integridade do dataframe resultante

```python
pd.concat([livros, livros_classicos], axis = 1, join = 'inner', keys = ['livros', 'livros_classicos'], names = ['tradicionais', 'classicos'], verify_integrity = True)
```
