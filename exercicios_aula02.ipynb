# **Exercícios**

***1) Qual setor possui mais funcionários?***

```python
funcionarios_e_setores = pd.merge(funcionarios, setores)
setores_func = funcionarios_e_setores['setor_nome'].value_counts(dropna=False, sort=True)

print(setores_func.iloc[[0]])
```

***2) Qual o cargo que possui maior quantidade de funcionários?***

***3) Qual o valor total de folha de pagamento para funcionários do setor de
compras?***

```python
setores_funcionarios = pd.merge(funcionarios, setores, how ='outer')
setores_funcionarios_cargos = pd.merge(setores_funcionarios, cargos, how ='outer')
soma_salarios = setores_funcionarios_cargos.groupby('setor_nome')['cargo_salario'].sum()
display(soma_salarios.iloc[[0]])
```

4) Qual é o funcionário mais velho do setor de marketing que possui o cargo de
analista?

```python
funcionarios_marketing = setores_funcionarios_cargos.loc[setores_funcionarios_cargos['setor_nome'] == 'Marketing']
marketing_e_analista = funcionarios_marketing.loc[funcionarios_marketing['cargo_nome'] == 'Analista']
ordenado_por_idade = marketing_e_analista.sort_values('funcionario_idade', ascending=False )

display(ordenado_por_idade.iloc[:3])
```

5) Qual o setor que tem mais funcionários de cargo de nível 1


```python
nivel_um = setores_funcionarios_cargos.loc[setores_funcionarios_cargos['cargo_nivel'] == 1]
nivel_por_setor = nivel_um.groupby('setor_nome')['cargo_nivel'].count()
ordenado_por_quantidade = nivel_por_setor.sort_values(ascending=False)

display(ordenado_por_quantidade.iloc[[0]])
```

6) Qual o nome dos funcionários com maior salário no setor de desenvolvimento?

```python
funcionarios_producao = setores_funcionarios_cargos.loc[setores_funcionarios_cargos['setor_nome'] == 'Desenvolvimento']
nomes_funcionarios = funcionarios_producao.loc[funcionarios_producao['cargo_salario'] == funcionarios_producao['cargo_salario'].max()][['funcionario_nome']]

pd.set_option('display.max_rows', 85)

display(nomes_funcionarios)
```

7) Qual é o funcionário mais novo no setor de compras e qual o mais velho?

```python
funcionarios_compras = setores_funcionarios_cargos.loc[setores_funcionarios_cargos['setor_nome'] == 'Compras']

funcionario_mais_velho = funcionarios_compras.loc[funcionarios_compras['funcionario_idade'] == funcionarios_compras['funcionario_idade'].max()][['funcionario_nome', 'funcionario_idade']]
funcionario_mais_novo = funcionarios_compras.loc[funcionarios_compras['funcionario_idade'] == funcionarios_compras['funcionario_idade'].min()][['funcionario_nome', 'funcionario_idade']]

display('>>>Lista com funcionários mais velhos:<<<') 
display(funcionario_mais_velho)
display('>>>Lista com funcionários mais novos:<<<') 
display(funcionario_mais_novo)
```
