3) Faça um comparativo entre os tipos de pokémon e as gerações,
evidencie em cada tipo quais gerações possuem mais de 50
pokémons. Dica: use pivoteamento.

```python
tipos_e_geracoes = pd.pivot_table(
    uniao_dataframes, 
    index=['pokemon_tipo_1_nome'],
    columns=['pokemon_geracao'], 
    aggfunc= {'pokemon_geracao': np.sum},
    fill_value = 0
    )

maior_50 = tipos_e_geracoes[tipos_e_geracoes['pokemon_geracao']>50]
maior_50 = maior_50[maior_50.notnull().any(axis=1)].fillna('-')

display(maior_50)
```

4) Faça um comparativo entre as gerações e a coluna de pokémons
lendários para verificar quantos pokémons existem com velocidade
acima de 85. Dica: use pivoteamento.

```python
geracao_e_lendarios = uniao_dataframes[uniao_dataframes['pokemon_velocidade']>85]

geracao_e_lendarios = pd.pivot_table(
    geracao_e_lendarios, 
    index=['pokemon_lendario'],
    columns=['pokemon_geracao'],
    aggfunc={'pokemon_geracao': len}
)

display(geracao_e_lendarios)
```

5) Plote um gráfico mostrando o valor somado da coluna
pokemon_total por tipo de pokemon, contando apenas os
pokemons que possuem defesa menor que 50.

```python
defesa_menor_50 = uniao_dataframes[uniao_dataframes['pokemon_defense']<50]
soma_pokemon_total = defesa_menor_50.groupby('pokemon_tipo_1_nome')['pokemon_total'].sum()

pd.options.plotting.backend = "plotly"

fig = px.bar(soma_pokemon_total, width=800, height=600, color_discrete_sequence=px.colors.qualitative.T10, template="plotly_dark")
fig.update_layout(title={
    'text' : 'Soma da coluna "pokemon_total" X tipo do pokemon.',
    'y': 0.93,
    'x': 0.5
})
fig.update_yaxes(title='Soma pokemon_total')
fig.update_xaxes(title='Tipo Pokemon')
fig.show()
```
